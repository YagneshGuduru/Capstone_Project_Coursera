```python
# Function to calculate the number of occurances of each value in a column.
def getValueCounts(column_name):
    return df[column_name].value_counts()

unique_items = df.nunique()
for i, v in unique_items.items():
    if 1 < v < 15:
        print(getValueCounts(i))
        print("-----")
```

###### Creating landing outcome label from outcome column.

```python
# Occurance of values in Outcome column.
landing_outcomes = df['Outcome'].value_counts()
type(landing_outcomes)
```

```python
# Extract the labels from landing_outcomes.
for i, l in enumerate(landing_outcomes.keys()):
    print(i,l)
```

```python
# Create a set of outcomes where second stage didn't land successfully.
bad_outcomes = set(landing_outcomes.keys()[[1,3,5,6,7]])
bad_outcomes
```

```python
# Create landing outcome label(Class).
landing_class = [0 if i in bad_outcomes else 1 for i in df['Outcome']]
```

```python
# Add new column CLASS to the dataframe.
df['Class'] = landing_class
```

```python
# df.head()
```

```python
# Check for the success rate of launches.
df['Class'].mean()
```

```python
# Load the new DataFrame into a CSV file.
df.to_csv("dataset_part_2.csv", index = False)
```
